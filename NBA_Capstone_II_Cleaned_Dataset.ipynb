```python
import pandas as pd #Importing the libraries that I will use to clean the dataset
import numpy as np
```

```python
nbaplayers = pd.read_csv("nbaplayers.csv")
```

```python
print(nbaplayers.head()) #Inspecting the columns, and the column heads are ugly
```

```python
nbaplayers.rename(columns = {'player_name':'Player', 'team_abbreviation':'Team', 'age':'Age','player_height':'Height(IN)', 'player_weight':'Weight(LBS)', 'college':'College', 'country':'Country', 'draft_year':'Draft Year', 'draft_round':'Draft Round', 'draft_number':'Draft Number', 'gp': 'GP', 'pts':'PPG', 'reb':'RPG', 'ast':'APG', 'net_rating':'Net Rating', 'oreb_pct': 'Offensive Rebound %', 'dreb_pct':'Defensive Rebound %', 'usg_pct':'Usage %', 'ts_pct':'True Shooting %', 'ast_pct':'Assist %', 'season':'Season'}, inplace=True)
```

```python
print(nbaplayers.columns) #Inspecting the renamed columns
```

```python
nbaplayers.isnull()
```

```python
nbaplayers.isnull().any() #Checking for null values in the dataset and there are none.
```

```python
nbaplayers.rename({'Unnamed: 0':'garbage'}, axis="columns", inplace=True) #renaming a column to drop
```

```python
print(nbaplayers.columns) #
```

```python
nbaplayers.drop(columns=['garbage']) #dropping the unnamed column from the dataset
```

```python
print(nbaplayers.columns)
```

```python
nbaplayers.rename(columns = {'Height':'Height(IN)', 'Weight':'Weight(LBS)'}, inplace=True)
```

```python
print(nbaplayers.columns) #remnamed the heigh and weight columns to indicate the calculated measure
```

```python
nbaplayers['Height(Inches)'] = (nbaplayers['Height(IN)'] / 2.54).round() #convert height in cm to in
```

```python
print(nbaplayers['Height(Inches)'])
```

```python
nbaplayers['Weight(Pounds)'] = (nbaplayers['Weight(LBS)'] * 2.2046).round() #converting weight in kg to lbs
```

```python
print(nbaplayers['Weight(Pounds)'])
```

```python
nbaplayers.drop(columns=['Height(IN)', 'Weight(LBS)']) #dropping the old weight and pounds columns
```

```python
nbaplayers.round({'Offensive Rebound %': 2, 'Defensive Rebound %': 2, 'Usage %': 2, 'True Shooting %': 2, 'Assist %': 2})
# rounding down some percentage statistics to make them easier to read
```

```python
nbaplayers.drop(['garbage', 'Height(IN)', 'Weight(LBS)'], axis=1, inplace=True) #dropping columns agin
```

```python
print(nbaplayers.head())
```

```python
nbaplayers = nbaplayers[['Player', 'Team', 'Age', 'Height(Inches)', 'Weight(Pounds)', 'College', 'Country', 'Draft Year', 'Draft Round', 'Draft Number', 'GP', 'PPG', 'RPG', 'APG', 'Net Rating', 'Offensive Rebound %', 'Defensive Rebound %', 'Usage %', 'True Shooting %', 'Assist %', 'Season']]
#reordering the columns
```

```python
nbaplayers.to_csv("Cleaned_Capstone_II_Dataset.csv") #saving the csv file after cleaning it
```
